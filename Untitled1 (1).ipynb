This is performing and importing the dataset for the visualization BY EDA and the model Training.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

sns.set(style="whitegrid")  

We, begin by loading the dataset and checking its shape, sample rows, missing values, and summary statistics  .


```python
df = pd.read_csv("bodyfat.csv")

print("Shape of dataset:", df.shape)
display(df.head())
display(df.isnull().sum())
display(df.describe())
```


Let's visualize how strongly features relate to each other using a heatmap.


```python
plt.figure(figsize=(15, 10))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm", square=True, linewidths=0.5)
plt.title(" Correlation Matrix of Features", fontsize=16)
plt.show() 
```

In this column the cell is running the code for the scaler.fit Transform .

```python
X = df.drop(columns=["BodyFat"])
y = df["BodyFat"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)       
plt.show() 
```

This cell is performing the  model to predict the X test and mean Squared error , for the better testing results 

```python
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f" Mean Squared Error (MSE): {mse:.4f}")
print(f" R-squared (R): {r2:.4f}") 
```

This Cell will show the graph of Actual and the predicted Body fat%

```python
results_df = pd.DataFrame({'Actual': y_test.values, 'Predicted': y_pred})
results_df_sorted = results_df.sort_values(by='Actual').reset_index(drop=True)

plt.figure(figsize=(12, 6))
plt.plot(results_df_sorted['Actual'], label='Actual', linewidth=2)
plt.plot(results_df_sorted['Predicted'], label='Predicted', linewidth=2, linestyle='--')
plt.title(" Actual vs Predicted Body Fat %", fontsize=16)
plt.xlabel("Sample Index")
plt.ylabel("Body Fat %")
plt.legend()
plt.grid(True)
plt.show() 
```

This cell will run the code of the actual and the predicted Scatter 

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Actual', y='Predicted', data=results_df, s=70, color='navy', edgecolor='white')
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.title(" Actual vs Predicted Scatter", fontsize=14)
plt.xlabel("Actual Body Fat %")
plt.ylabel("Predicted Body Fat %")
plt.grid(True)
plt.show() 
```

This cell shows the Residual distribution 

```python
residuals = y_test - y_pred

plt.figure(figsize=(10, 6))
sns.histplot(residuals, kde=True, bins=30, color="darkmagenta")
plt.axvline(0, color='black', linestyle='--')
plt.title(" Residual Distribution", fontsize=14)
plt.xlabel("Prediction Error (Residual)")
plt.ylabel("Frequency")
plt.show() 
```

In this cell it will show the feature importance and Model Coefficients.

```python
coef_df = pd.DataFrame({
    "Feature": df.drop(columns=["BodyFat"]).columns,
    "Coefficient": model.coef_
}).sort_values(by="Coefficient", key=abs, ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x="Coefficient", y="Feature", data=coef_df, palette="viridis")
plt.title("Feature Importance (Model Coefficients)", fontsize=14)
plt.grid(True)
plt.show() 
```

Pairplot for better visualization 

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning) 
top_features = df.corr()['BodyFat'].abs().sort_values(ascending=False).index[1:6]
sns.pairplot(df[top_features.tolist() + ['BodyFat']], corner=True)
plt.suptitle("Pairplot of Top Correlated Features with BodyFat", y=1.02)
plt.show() 
```

Prediction error plot 

```python
plt.figure(figsize=(10, 5))
errors = y_test - y_pred
sns.scatterplot(x=y_pred, y=errors, alpha=0.7, s=60)
plt.axhline(0, linestyle='--', color='red')
plt.title("Prediction Error Plot")
plt.xlabel("Predicted Body Fat %")
plt.ylabel("Prediction Error")
plt.grid(True)
plt.show()
```

```python
import scipy.stats as stats
import statsmodels.api as sm

sm.qqplot(residuals, line='s')
plt.title("QQ Plot of Residuals")
plt.grid(True)
plt.show()
```
